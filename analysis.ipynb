```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
dir_path = 'Latencies/tp_10k'
lat_file_paths = [file for file in os.listdir(dir_path) if file.endswith('.csv')]
```

```python
# df = pd.read_csv('/home/yash/Desktop/Clg/7th_Sem/SDS/Assignment_3/AIR_regex/Throughputs/throughputs_DF1_10k.csv', header = None, names = ['WID', 'Throughput'])
plt.figure(figsize=(10, 6))

for csv_file in lat_file_paths:
    file_path = os.path.join(dir_path, csv_file)

    df = pd.read_csv(file_path, header=None, names=['Window', 'Latencies'])

    window_numbers = df['Window']
    lat_values = df['Latencies']

    plt.plot(window_numbers, lat_values, linestyle='-', label=csv_file)

plt.legend(loc = 'best')
plt.xlabel('WID')
plt.ylabel('Latencies')
plt.title('Latencies vs WID (tp = 10k)')


plt.show()
```
